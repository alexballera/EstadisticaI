###**Librerías**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
import math
from IPython.display import display, Math
import yfinance as yf
```

###Ejemplo

Se extraen los datos de cotización y cantidades para Google, Apple y Amazon entre el 16/6/2023 y el 30/6/2023

```python
data = pd.DataFrame(yf.download(['GOOG', 'AAPL', 'AMZN'], start='2023-06-16', end='2023-07-01')).reset_index()
data.columns = data.columns.map(''.join)
data
```

```python
data=data.loc[:, data.columns.isin(['Date','CloseAAPL','CloseAMZN','CloseGOOG','VolumeAAPL','VolumeAMZN','VolumeGOOG'])]
data
```

### Índice de precio simple

Construcción de los índice de precios simple con base 16/6/2023

```python
data['I_APPL']=round((data['CloseAAPL']/data['CloseAAPL'][0])*100,2)
data['I_AMZN']=round((data['CloseAMZN']/data['CloseAMZN'][0])*100,2)
data['I_GOOG']=round((data['CloseGOOG']/data['CloseGOOG'][0])*100,2)
data
```

### Índice agregado simple

Continuando con el ejemplo anterior calcular el índice de precio la cartera para el 30/6/2023.

```python
display(Math(r'I_{30/6/2023;16/6/2023}^{Ag.Sim.}= {\sum_{i=1}^n p_{k;t} \over \sum_{i=1}^n p_{k;0}}*100 ='+str(round(((data['CloseAAPL'][9]+data['CloseAMZN'][9]+data['CloseGOOG'][9]) / (data['CloseAAPL'][0]+data['CloseAMZN'][0]+data['CloseGOOG'][0]))*100,2))))
```

### Índice agregado ponderado

Continuando con el ejemplo anterior calcular el índice de la cartera ponderado para el
30/6/2023. Para ello suponer que el titular de la cartera compra 20 acciones de APPL,
50 de GOOGL y 80 de AMZN.

```python
display(Math(r'I_{30/6/2023;16/6/2023}^{Ag.Pond.}= {\sum_{i=1}^n q^k*p_{k;t} \over \sum_{i=1}^n q^k*p_{k;0}}*100 ='+str(round(((20*data['CloseAAPL'][9]+ 80*data['CloseAMZN'][9]+ 50*data['CloseGOOG'][9]) / (20*data['CloseAAPL'][0]+ 80*data['CloseAMZN'][0]+ 50*data['CloseGOOG'][0]))*100,2))))
```

###Índice de Laspeyre

Cálculo del Índice de Laspeyre para el 30/6/2023

```python
display(Math(r'IL_{30/6/2023;16/6/2023}= {\sum_{i=1}^n q_{i;0}*p_{i;t} \over \sum_{i=1}^n q_{i;0}*p_{i;0}}*100 ='+str(round(((data['VolumeAAPL'][0]*data['CloseAAPL'][9]+ data['VolumeAMZN'][0]*data['CloseAMZN'][9]+ data['VolumeGOOG'][0]*data['CloseGOOG'][9])/(data['VolumeAAPL'][0]*data['CloseAAPL'][0]+ data['VolumeAMZN'][0]*data['CloseAMZN'][0]+ data['VolumeGOOG'][0]*data['CloseGOOG'][0]))*100,2))))
```

###Índice de Paschee

Cálculo del Índice de Laspeyre para el 30/6/2023

```python
display(Math(r'IP_{30/6/2023;16/6/2023}= {\sum_{i=1}^n q_{i;t}*p_{i;t} \over \sum_{i=1}^n q_{i;t}*p_{i;0}}*100 ='+str(round(((data['VolumeAAPL'][9]*data['CloseAAPL'][9]+ data['VolumeAMZN'][9]*data['CloseAMZN'][9]+ data['VolumeGOOG'][9]*data['CloseGOOG'][9])/(data['VolumeAAPL'][9]*data['CloseAAPL'][0]+ data['VolumeAMZN'][9]*data['CloseAMZN'][0]+ data['VolumeGOOG'][9]*data['CloseGOOG'][0]))*100,2))))
```

###Índice de Fisher

Cálculo del Índice de Fisher para el 30/6/2023

```python
IL=round(((data['VolumeAAPL'][0]*data['CloseAAPL'][9]+ data['VolumeAMZN'][0]*data['CloseAMZN'][9]+ data['VolumeGOOG'][0]*data['CloseGOOG'][9])/(data['VolumeAAPL'][0]*data['CloseAAPL'][0]+ data['VolumeAMZN'][0]*data['CloseAMZN'][0]+ data['VolumeGOOG'][0]*data['CloseGOOG'][0]))*100,2)
IP=round(((data['VolumeAAPL'][9]*data['CloseAAPL'][9]+ data['VolumeAMZN'][9]*data['CloseAMZN'][9]+ data['VolumeGOOG'][9]*data['CloseGOOG'][9])/(data['VolumeAAPL'][9]*data['CloseAAPL'][0]+ data['VolumeAMZN'][9]*data['CloseAMZN'][0]+ data['VolumeGOOG'][9]*data['CloseGOOG'][0]))*100,2)
display(Math(r'IF_{30/6/2023;16/6/2023}= \sqrt {IL_t*IP_t} ='+str(round((IL*IP)**0.5,2))))
```

###Cambio de base


Tomemos el "Índice de precios de GOOGL“ (con base 16/6) y el de AMZN (con base
23/6). Asumamos que queremos analizar en cual índice hubo mayores cambios.

```python
data2=data.loc[:, data.columns.isin(['Date','CloseAMZN','CloseGOOG'])]
data2
```

Calculamos el índice de precio de Google con base el 16/6 y el Amazon con base el 23/6

```python
data2['I_AMZN']=round((data2['CloseAMZN']/data2['CloseAMZN'][4])*100,2)
data2['I_GOOG']=round((data2['CloseGOOG']/data2['CloseGOOG'][0])*100,2)
data2
```

Realizamos el cambio de base para google al 23/6/2023

```python
data2['I_GOOG*']=round((data2['I_GOOG']/data2['I_GOOG'][4])*100,2)
data2
```

Ahora podemos calcular variaciones que sean comparables

```python
display(Math(r'\bigtriangleup IGOOG^*_{30/6/2023;29/6/2023}='+str(round(data2['I_GOOG*'][9]-data2['I_GOOG*'][8],2))))
print('')
display(Math(r'\bigtriangleup IAMZN_{30/6/2023;29/6/2023}='+str(round(data2['I_AMZN'][9]-data2['I_AMZN'][8],2))))
```

```python
display(Math(r'\%\bigtriangleup IGOOG^*_{30/6/2023;29/6/2023}='+str(round((data2['I_GOOG*'][9]-data2['I_GOOG*'][8])/data2['I_GOOG*'][8],2))))
print('')
display(Math(r'\%\bigtriangleup IAMZN_{30/6/2023;29/6/2023}='+str(round((data2['I_AMZN'][9]-data2['I_AMZN'][8])/data2['I_AMZN'][8],2))))
```
